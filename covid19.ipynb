```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
```

```python
confirmed = pd.read_csv('/content/time_series_covid19_confirmed_global.csv', sep = ',')
deaths = pd.read_csv('/content/time_series_covid19_deaths_global.csv', sep = ',')
```

```python
confirmed.fillna(value=0, inplace=True)
confirmed.head()
```

```python
deaths.fillna(value=0, inplace=True)
deaths.head()
```

```python
new_cols = list(confirmed.columns[:4]) + list(confirmed.columns[4:].map(lambda x: '{0:02d}.{1:02d}.{2:d}'.format(int(x.split(sep='/')[1]), int(x.split(sep='/')[0]), int(x.split(sep='/')[2]))))
```

```python
confirmed.columns = new_cols
deaths.columns = new_cols
```

```python
deaths.head()
```

```python
confirmed.columns.get_loc('01.03.22')
```

```python
confirmed.columns.get_loc('01.03.23')
```

```python
conf1 = confirmed.iloc[:, [0, 1, 2, 3]]
conf1.head()
```

```python
conf2 = confirmed.iloc[:, 773:1139]
conf2.head()
```

```python
confirmed1 = pd.concat([conf1, conf2], axis=1, join='inner')
```

```python
confirmed1.fillna(value=0, inplace=True)
confirmed1.head()
```

```python
de1 = deaths.iloc[:, [0, 1, 2, 3]]
de1.head()
```

```python
de2 = deaths.iloc[:, 773:1139]
de2.head()
```

```python
deaths1 = pd.concat([de1, de2], axis=1, join='inner')
```

```python
deaths1.fillna(value=0, inplace=True)
deaths1.head()
```

```python
confirmed_daily = confirmed1.copy()
confirmed_daily.iloc[:,4:] = confirmed_daily.iloc[:,4:].diff(axis=1)
deaths_daily = deaths1.copy()
deaths_daily.iloc[:,4:] = deaths_daily.iloc[:,4:].diff(axis=1)
```

```python
smooth_conf_daily = confirmed_daily.copy()
smooth_conf_daily.iloc[:,4:] = smooth_conf_daily.iloc[:,4:].rolling(window=8, min_periods=2, center=True, axis=1).mean()
smooth_conf_daily.iloc[:,4:] = smooth_conf_daily.iloc[:,4:].round(1)
```

```python
last_date = confirmed.columns[-1]
```

```python
confirmed_top = confirmed.iloc[:, [1, -1]].groupby('Country/Region').sum().sort_values(last_date, ascending = False).head(20)
countries = list(confirmed_top.index)
```

```python
conf_tot_ratio = confirmed_top.sum() / confirmed.iloc[:, 4:].sum()[-1]
```

```python
countries.append('Russia')
```

```python
l1 = 'Infected at ' + last_date + ' - ' + str(confirmed.iloc[:, 4:].sum()[-1])
l2 = 'Dead at ' + last_date + ' - ' + str(deaths.iloc[:, 4:].sum()[-1])
```

```python
fig, ax = plt.subplots(figsize = [16,6])
ax.plot(confirmed.iloc[:, 4:].sum(), '-', alpha = 0.6, color = 'orange', label = l1)
ax.plot(deaths.iloc[:, 4:].sum(), '-', alpha = 0.6, color = 'red', label = l2)

ax.legend(loc = 'upper left', prop=dict(size=12))
ax.xaxis.grid(which='minor')
ax.yaxis.grid()
ax.tick_params(axis = 'x', labelrotation = 90)
plt.title('COVID-19 in all countries. Top 20 countries consists {:.2%} of total confirmed infected cases.'.format(conf_tot_ratio[0]))
plt.show()
```

```python
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, normalize
```

```python
confirmed1 = confirmed1.drop('Country/Region', axis=1)
```

```python
confirmed1 = confirmed1.drop('Province/State', axis=1)
```

```python
confirmed1 = confirmed1.drop('Lat', axis=1)
```

```python
confirmed1 = confirmed1.drop('Long', axis=1)
```

```python
confirmed1.head()
```

```python
scaler = StandardScaler()
confirmed1_scaled = scaler.fit_transform(confirmed1)
confirmed1_normalized = normalize(confirmed1_scaled)
confirmed1_normalized = pd.DataFrame(confirmed1_normalized)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
confirmed1_principal = pca.fit_transform(confirmed1_normalized)
confirmed1_principal = pd.DataFrame(confirmed1_principal)
```

```python
confirmed2 = confirmed1_principal
clustering = AgglomerativeClustering().fit(confirmed2)
```

```python
clustering.labels_
```

```python
confirmed2.columns = ['C1', 'C2']
ac2 = AgglomerativeClustering(n_clusters = 2)

plt.figure(figsize =(6, 6))
plt.scatter(confirmed2['C1'], confirmed2['C2'], c = ac2.fit_predict(confirmed2), cmap ='rainbow')
plt.show()
```
